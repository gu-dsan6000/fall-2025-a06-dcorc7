```python
import os
import sys

# 1. Create necessary directories
os.system("mkdir -p data/raw data/output")

# 2. Download the dataset archive from the course S3 bucket
os.system("aws s3 cp s3://dsan6000-datasets/spark-logs/Spark.tar.gz data/raw/ --request-payer requester")

# 3. Extract the archive
os.chdir("data/raw/")
os.system("tar -xzf Spark.tar.gz")
os.chdir("../..")

# 4. Verify extraction
os.system("ls data/raw/application_* | head -10")

# 5. Create your personal S3 bucket
YOUR_NET_ID = "dc1524"
bucket_name = f"{YOUR_NET_ID}-assignment-spark-cluster-logs"
os.system(f"aws s3 mb s3://{bucket_name}")

# 6. Upload the data to S3
os.system(f"aws s3 cp data/raw/ s3://{bucket_name}/data/ --recursive --exclude '*.gz' --exclude '*.tar.gz'")

# 7. Verify upload
os.system(f"aws s3 ls s3://{bucket_name}/data/ | head -10")

# 8. Save your bucket name for later use
bashrc_entry = f"export SPARK_LOGS_BUCKET=s3://{bucket_name}\n"
with open(os.path.expanduser("~/.bashrc"), "a") as f:
    f.write(bashrc_entry)

# Reload environment variables
os.system("source ~/.bashrc")
```
